```python
# ? What are the location with fastest roaler coaster ?
```

```python
df.head()
```

```python
df['Location'].value_counts()
```

```python
df.query('Location != "other"').groupby('Location')['Speed_mph'].agg(['mean','count'])
```

```python
df.query('Location != "other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >=10')
```

```python
df.query('Location != "other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >=10') \
    .sort_values('mean')
```

```python
df.query('Location != "other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >=10') \
    .sort_values('mean') \
    .plot(kind='barh', figsize=(12,5), title = 'Average Count Speed by Location')
```

```python
df.query('Location != "other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >=10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12,5), title = 'Average Count Speed by Location')
```

```python
ax = df.query('Location != "other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >=10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12,5), title = 'Average Count Speed by Location')
ax.set_xlabel('Avrage Coaster speed')
plt.show()
```
